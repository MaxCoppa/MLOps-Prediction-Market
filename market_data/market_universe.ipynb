```python
import requests
import pandas as pd
@
API_URL = "https://api.elections.kalshi.com/trade-api/v2"

def get_tickers(series_ticker, max_pages=100):
    all_markets = []
    cursor = None
    
    # Iterate through pages (safety limit of 5 pages = 5000 markets)
    for _ in range(max_pages):
        params = {"status": "open", "limit": 1000, "series_ticker":series_ticker}
        if cursor:
            params["cursor"] = cursor
            
        try:
            response = requests.get(f"{API_URL}/markets", params=params, timeout=10)
            response.raise_for_status()
            data = response.json()
            
            batch = data.get("markets", [])
            if not batch:
                break
            
            all_markets.extend(batch)
            cursor = data.get("cursor")
            if not cursor:
                break
        except Exception as e:
            print(f"Error: {e}")
            break

    df = pd.DataFrame(all_markets)
    return df

# Run
bets = get_tickers("KXFEDCHAIRNOM")
```
